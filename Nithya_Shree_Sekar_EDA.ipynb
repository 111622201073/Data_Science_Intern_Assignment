```python
#loading dataset
import pandas as pd
```

```python
customer=pd.read_csv('Customers.csv')
print(customer.head(10))
```

```python
product=pd.read_csv('Products.csv')
product.head(10)
```

```python
transaction=pd.read_csv('Transactions.csv')
transaction.head(10)
```

```python
#data preprocessing
merged=pd.merge(customer,transaction,on="CustomerID",how="inner")
new_merged=pd.merge(merged,product,on=["ProductID","Price"],how="inner")
print(new_merged)
```

```python
new_merged.drop_duplicates()
```

```python
#data analysis and visualization
new_merged.describe()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,6))
plt.scatter(new_merged['ProductID'],new_merged['Price'],color='green')
plt.title('Distribution of price andTotal Transaction Values')
plt.xlabel('Total Transaction Value')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.bar(new_merged['Quantity'],new_merged['TotalValue'])
plt.title('Distribution of quantity overTotal Transaction Values')
plt.ylabel('Total Transaction Values')
plt.xlabel('Quantity')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.plot(new_merged['Region'],new_merged['Quantity'])
plt.title('Distribution of Quantity over Region')
plt.xlabel('Region')
plt.ylabel('Quantity')
plt.show()
```

```python
#outlier detection
def identify_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    outliers = column[(column < (Q1 - 1.5 * IQR)) | (column > (Q3 + 1.5 * IQR))]
    return outliers
price_outliers=identify_outliers(new_merged['Price'])
quantity_outliers=identify_outliers(new_merged['Quantity'])
total_value_outliers=identify_outliers(new_merged['TotalValue'])
print(f'Price Outliers:\n{price_outliers}')
print(f'Quantity Outliers:\n{quantity_outliers}')
print(f'Total Value Outliers:\n{total_value_outliers}')
```

```python
def plot_distribution(column, column_name):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.hist(column,color="pink")
    plt.title(f'Distribution of {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
plot_distribution(new_merged['Price'], 'Price')
plot_distribution(new_merged['Quantity'], 'Quantity')
plot_distribution(new_merged['TotalValue'], 'Total Value')
```
